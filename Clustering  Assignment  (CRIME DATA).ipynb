Hierarchical Clustering

```python
#import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch      # to build dendrogram and build the plotting
from sklearn.cluster import AgglomerativeClustering

import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
crime = pd.read_csv("crime_data.csv")
```

```python
crime
```

```python
crime.info()
```

```python
#Customized Normalization Function 
def norm_func(i):            
    x = (i-i.min())/(i.max()-i.min())              #(Xi-min)/range             range = max-min
    return(x)
```

```python
#(i-i.min())/(i.max()-i.min())
crime['Murder'].min(),crime['Murder'].max()
```

```python
(13.2-0.8)/(17.4-0.8)
```

```python
#Normalized data frame (considering the numerical part of data)
df_norm = norm_func(crime.iloc[:,1:]) 
```

```python
df_norm
```

```python
#Create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm,method='single'))        #single linkage 
```

```python
#create clusters , suppose got input from customer that go for 2 clusters
hc = AgglomerativeClustering(n_clusters=2,affinity = 'euclidean',linkage = 'single')
```

```python
hc
```

```python
#save clusters for chart
y_hc = hc.fit_predict(df_norm)
Clusters = pd.DataFrame(y_hc,columns=['Clusters'])  #append those no. of cluster numbers create dataframe
```

```python
y_hc
```

```python
Clusters
```

```python
#Now let us map this cluster membership to the data points
crime['h_clustered'] = Clusters
```

```python
crime
```

```python
crime1 = crime.sort_values("h_clustered")       #sorting 
crime1.iloc[:,[0,5]]
```

Kmeans

```python
#dont know the no. of clusters . so lets use Kmeans and elbow method to choose this number of optional clusters
from sklearn.cluster import KMeans
```

```python
crime1 = pd.read_csv("crime_data.csv")
```

```python
crime1
```

```python
#Normalization/Standardization Function 
from sklearn.preprocessing import StandardScaler            #built in  function for standardization
scaler = StandardScaler()        
scaled_crime1_df = scaler.fit_transform(crime1.iloc[:,1:])
scaled_crime1_df
```

```python
wcss  = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,random_state=0)
    kmeans.fit(scaled_crime1_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number Of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
wcss
```

```python
kmeans = KMeans(n_clusters = 4,random_state=0)                   #pass no. of clusters as 4
predict=kmeans.fit_predict(scaled_crime1_df)                     #assign cluster numbers to crime1
```

```python
predict
```

```python
clusters_new = KMeans(n_clusters = 3,random_state=0)                   #pass no. of clusters as 3
new_predict=clusters_new.fit_predict(scaled_crime1_df)    
```

```python
new_predict
```

```python
clusters_new.labels_
```

```python
#Assign clusters to the data set
#crime1['clusterid_new'] = clusters_new.labels_
crime1['clusterid_new'] = new_predict
```

```python
crime1
```

```python
crime1.groupby('clusterid_new').agg(['mean']).reset_index()
```
